#### 3.2. Добавить день недели, месяц и год публикации объявления

```python
# Добавляем день недели
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday+1
# Для удобства прибавляю 1, т.к. эта функция ведет счет с 0

# добавляем месяц
data['month_exposition'] = data['first_day_exposition'].dt.month

# добавляем год
data['year_exposition'] = data['first_day_exposition'].dt.year
```

#### 3.3. Этажность разбить на 3 группы (1-й, последний, другой)

```python
#Используем функцию для категоризации по этажам:
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
# Создадим cтолбец с категориями:
data['floor_category'] = data.apply(floor_category, axis=1)
data['floor_category'].value_counts()
```
